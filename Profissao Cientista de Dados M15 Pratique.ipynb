##### Podemos ver que tanto a coluna de pagamento mensal quanto a coluna de total pagos tem outlier principalmente 

```python

fig1 = px.box(df, y='monthly_charges', title='Boxplot de Monthly Charges')
fig1.show()

# podemos ver nesse grafico mais claramente que existem outliers na coluna de pagamento mensal
```

```python
def iqr_method(data_frame,coluna):

    q1 = data_frame[coluna].quantile(0.25)
    q3 = data_frame[coluna].quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    return limite_inferior, limite_superior

limite_inferior, limite_superior = iqr_method(df, 'monthly_charges')
df = df[(df['monthly_charges'] >= limite_inferior) & (df['monthly_charges'] <= limite_superior)]

df.describe()
```

```python

fig1 = px.box(df, y='monthly_charges', title='Boxplot de Monthly Charges')
fig1.show()
```

```python


limite_inferior_1, limite_superior_1 = iqr_method(df, 'total_charges')
limite_inferior_2, limite_superior_2 = iqr_method(df, 'tenure')
df = df[(df['total_charges'] >= limite_inferior_1) & (df['total_charges'] <= limite_superior_1)]
df = df[(df['tenure'] >= limite_inferior_2) & (df['tenure'] <= limite_superior_2)]
```

#### Relação de Churn por Genero

```python
#Quem é mais propenço a dar churn , homens ou mulheres?
fig = px.histogram(df, x='churn' , color='gender', barmode='stack')

fig.update_layout(title_text='Relação de Churn por Genero',
                   title_x=0.5, xaxis_title='churn',
                     yaxis_title='Contagem' ,
                       legend_title='gender')
fig.show()
```

```python
df_grouped = df.groupby(['gender', 'churn']).size().reset_index(name='count')
total_counts = df_grouped.groupby('gender')['count'].transform('sum')
df_grouped['percentage'] = (df_grouped['count'] / total_counts) * 100

fig = px.bar(df_grouped, x='churn', 
             y='percentage', 
             color='gender',
               barmode='stack', 
               labels={'Churn':'churn','percent':'percentage','Genero':'gender'}
                 )
fig.update_layout(title_text='Relação de Churn por Genero',
                   title_x=0.5, xaxis_title='churn',
                     yaxis_title='Contagem' ,
                       legend_title='gender')
fig.show()
```

Isso mostra que, mesmo os dados sendo balanciados, o número de churn é praticamente igual mas a taxa de saída é maior entre mulheres.

#### Relação de Churn por Idosos

```python
# Será que idosos tem mais chance de dar churn que os mais jovens?
df_grouped = df.groupby(['senior', 'churn']).size().reset_index(name='count')
total_counts = df_grouped.groupby('senior')['count'].transform('sum')
df_grouped['percentage'] = (df_grouped['count'] / total_counts) * 100

fig = px.bar(df_grouped, x='churn', 
             y='percentage', 
             color='senior',
               barmode='stack', 
               labels={'Churn':'churn','percent':'percentage','Senior':'senior'}
                 )
fig.update_layout(title_text='Relação de Churn por Idosos',
                   title_x=0.5, xaxis_title='churn',
                     yaxis_title='Contagem' ,
                       legend_title='senior')
fig.show()
```

podemos ver que idosos tendem a dar mais churn que pessoas jovens isso sugere que ser idoso é um fator de risco para churn (talvez por menor engajamento com tecnologia, custo elevado, ou falta de adaptação aos serviços).

#### Relação entre Estado Civil, Dependentes e Churn

```python
#será que solteiros com dependentes tendem a dar mais churn?
df['married_dependents'] = df['married'] + '_' + df['dependents']
df_grouped = df.groupby(['married_dependents', 'churn']).size().reset_index(name='count')
total_counts = df_grouped.groupby('married_dependents')['count'].transform('sum')
df_grouped['percentage'] = (df_grouped['count'] / total_counts) * 100
fig = px.bar(df_grouped, x='married_dependents',
                y='percentage',
                color='churn',
                barmode='stack',
                labels={'married_dependents':'Estado Civil e Dependentes',
                        'percentage':'Percentual (%)',
                        'churn':'Churn'}
                        , title='Relação entre Estado Civil, Dependentes e Churn'
                )
fig.show()
```

o numero de churn entre pessoas casadas e solteiras sem dependentes é maior que a de pesssoas casadas e solteiras com dependente , o que pode estar relacionado ao uso por pais e maes para entreter seus filhos atravez de conteudo infantil na internet, não é tão conclusivo mais e algo a se pensar

#### Relação entre Número de Serviços e Churn

```python
#sera que pessoas com mais de 1 serviço tendem a dar mais churn?

df['internet_service'] = df['internet_service'].apply(lambda x: 'NO' if x == 'NO' else 'YES')

df['total_services'] = df[['phone_service', 'internet_service', 'online_security', 'tech_support', 'streaming_tv']].transform(lambda x: x == 'YES').sum(axis=1)

# Verificar a distribuição dos serviços
print("Distribuição do número de serviços:")
print(df['total_services'].value_counts().sort_index())

# Agrupar e plotar
df_grouped = df.groupby(['total_services', 'churn']).size().reset_index(name='count')
total_counts = df_grouped.groupby('total_services')['count'].transform('sum')
df_grouped['percentage'] = (df_grouped['count'] / total_counts) * 100

fig = px.bar(df_grouped, 
             x='total_services',
             y='percentage',
             color='churn',
             barmode='stack',
             labels={'total_services':'Número de Serviços',
                     'percentage':'Percentual (%)',
                     'churn':'Churn'},
             title='Relação entre Número de Serviços e Churn')
fig.show()
```

podemos notar no grafico que pessoas com 2 ou 3 produtos tem mais chance de deixar a marque que pessoas com 1, 4 e 5 produtos
