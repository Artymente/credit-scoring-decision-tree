# 3) Aplique o algortimo da árvore de decisão aos dados de treinamento, utilizando critério de Gini e random state = 0.
Traga a acurácia para o modedlo com os dados de treino.

```python
score_tree = DecisionTreeClassifier(criterion='gini', random_state=0)
score_tree.fit(x_train, y_train)
```

```python
previcao = score_tree.predict(x_train)
previcao
```

```python
print(y_train)
```

```python
acuracia_traino = accuracy_score(y_train, previcao)
print(f"Acurácia no treino: {acuracia_traino}")
```

```python
relatorio = classification_report(y_train, previcao)
print(relatorio)
```

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

```python
previsao = score_tree.predict(x_test)
print(previsao)
```

```python
relatorio = classification_report(y_test, previsao)
print(relatorio)
```

```python
c_matrix = confusion_matrix(y_test, previsao)

plt.figure(figsize=(6,5))
sns.heatmap(c_matrix,
            annot=True,
            fmt="d",
            cmap="Blues",
            xticklabels=["Baixo", "Médio", "Alto"],
            yticklabels=["Baixo", "Médio", "Alto"])

plt.xlabel("Predito")
plt.ylabel("Real")
plt.title("Matriz de Confusão - Arvore de decisão")
plt.show()
```

O modelo apresentou, na base de teste, uma performance muito próxima da base de treino, ambas com resultados próximos de 100%. Esse comportamento levanta a possibilidade da ocorrência de overfitting, já que um bom desempenho no teste indica capacidade preditiva, mas apenas a estabilidade em múltiplas amostras confirma a real generalização do modelo apesar dessa ressalva, o modelo demonstrou um desempenho muito positivo na base de teste. Observa-se uma diferença de aproximadamente 14 pontos percentuais na métrica de precision para a categoria de score “médio”, indicando a presença de alguns falsos positivos nessa classe. Já para a categoria de score alto, a diferença observada foi de cerca de 3 pontos percentuais no recall, mostrando que o modelo deixou de identificar uma pequena parcela dos casos reais dessa classe de forma geral, os resultados indicam que o modelo conseguiu capturar bem os padrões dos dados, inclusive em um cenário de desbalanceamento entre classes. No entanto, devido à alta complexidade do modelo e à proximidade excessiva entre os desempenhos de treino e teste, recomenda-se a aplicação de técnicas adicionais, como validação cruzada e controle de complexidade, para confirmar a robustez e a capacidade de generalização do modelo.
